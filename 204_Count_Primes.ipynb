## 204. Count Primes

#### Given an integer n, return the number of prime numbers that are strictly less than n.

Example 1:

Input: n = 10
Output: 4
Explanation: There are 4 prime numbers less than 10, they are 2, 3, 5, 7.

Example 2:

Input: n = 0
Output: 0

Example 3:

Input: n = 1
Output: 0
 

Constraints:

0 <= n <= 5 * 106

```python
class Solution:
    def countPrimes(self, n: int) -> int:
        
        if n<=1:
            return 0
        
        li = [False,False]+[True]*(n-2)
        
        for i in range(2,int(sqrt(n))+1):
            for j in range(i*i,n,i):
                li[j] = False
        
        return sum(li)      
```

```python
class Solution:
    def countPrimes(self, n: int) -> int:
        
        if n <= 2:
            return 0

        is_prime = [True] * n
        is_prime[0] = False
        is_prime[1] = False

        for i in range(2, int(n ** 0.5) + 1):
            if is_prime[i]:
                for j in range(i * i, n, i):
                    is_prime[j] = False

        return sum(is_prime)
```

```python
class Solution:
    def countPrimes(self, n: int) -> int:
        
        if n <= 2: 
            return 0
        
        is_prime = [False,False] + [True] * (n-2)
        
        p = 2
        
        while p*p < n:
            if is_prime[p]:
                for i in range(p*p,n,p):
                    is_prime[i] = False
            p += 1
        
        return sum(is_prime)        
```
