## 680. Valid Palindrome II

Given a string s, return true if the s can be palindrome after deleting at most one character from it.

Example 1:

Input: s = "aba"

Output: true

Example 2:

Input: s = "abca"

Output: true

Explanation: You could delete the character 'c'.

Example 3:

Input: s = "abc"

Output: false
 

Constraints:

1 <= s.length <= 105
s consists of lowercase English letters.

```python
class Solution:
    def validPalindrome(self, s: str) -> bool:
        
        n = len(s)
        left = 0
        right = n-1

        def pal(st):
            return st==st[::-1]

        while left < right:
            
            if s[left] == s[right]:
                left += 1
                right -= 1
            
            else:
                left_half = s[:left]+s[left+1:]
                right_half = s[:right]+s[right+1:]
                return pal(left_half) or pal(right_half)

        return True
```
