## 5. Longest Palindromic Substring

Given a string s, return the longest palindromic substring in s.

Example 1:

Input: s = "babad"

Output: "bab"

Explanation: "aba" is also a valid answer.

Example 2:

Input: s = "cbbd"

Output: "bb"
 

Constraints:

1 <= s.length <= 1000

s consist of only digits and English letters.

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        if len(s) <= 1:
            return s
        
        Max_Len=1
        Max_Str=s[0]
        
        for i in range(len(s)-1):
            
            for j in range(i+1,len(s)):
            
                if j-i+1 > Max_Len and s[i:j+1] == s[i:j+1][::-1]:
                    Max_Len = j-i+1
                    Max_Str = s[i:j+1]

        return Max_Str
```

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        n = len(s)
        res = ""

        for i in range(n):
            
            st = end = i
            
            while st >= 0 and end < n and s[st] == s[end]:
                st -= 1
                end += 1
            
            temp = s[st+1:end]
            
            if len(temp) > len(res):
                res = temp

            st, end = i, i+1
            
            while st >= 0 and end < n and s[st] == s[end]:
                st -= 1
                end += 1
            
            temp = s[st+1:end]
            
            if len(temp) > len(res):
                res = temp

        return res
```
